#Findings

```python
categorical_columns=['Gender','Embarked','Name','Ticket','Cabin']
ds=pd.get_dummies(data=ds,columns=categorical_columns,drop_first=True)
ds.head()
```

```python
Q1=ds['Age'].quantile(0.25)
Q3=ds['Age'].quantile(0.75)
IQR=Q3-Q1
ds=ds[~((ds['Age']<(Q1-1.5*IQR))|(ds['Age']>(Q3+1.5*IQR)))]
print(ds['Age'].describe())
```
